```python
import numpy as nu
import math as ma
x=nu.random.uniform(-10,10)
eta=0.01
max=100
i=0
while i<max:
  slopex=4*ma.pow(x,3)+ 6*x
  stepxlen=-eta*slopex
  x=x+stepxlen
  i=i+1
min=ma.pow(x,4)+3*ma.pow(x,2)+10
print("Control variable x= ",x)
print("objective function = ",min)  
```
