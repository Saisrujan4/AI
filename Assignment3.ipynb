```python
import numpy as nu
import math as ma
x=nu.random.uniform(-10,10)
y=nu.random.uniform(-10,10)
eta=0.01
max=100
i=0
while i<max:
  slopex=6*x
  slopey=-5*ma.exp(-y)
  steplenx=-eta*slopex
  stepleny=-eta*slopey
  x=x+steplenx
  y=y+stepleny
  i=i+1
min=3*ma.pow(x,2)+5*ma.exp(-y)+10
print("x=",x)
print("y=",y)
print("objective function = ",min)
```
